## Quiz Questions

### Question 1

```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
# upload a dataset from seaborn
penguins = seaborn.load_dataset('penguins')
```

```python
%config InlineBackend.figure_format = 'retina'
```

### Question 2

```python
# create a scatter plot in Seaborn for variables x and y
seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
pyplot.show()
```

### Question 3

```python
# create a line plot in Seaborn for variables x and y
seaborn.lineplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
pyplot.show()
```

### Question 4

```python
# create a histogram in Seaborn for a variable x
seaborn.histplot(data=penguins, x='bill_length_mm')
pyplot.show()
```

### Question 5

```python
# create a stacked histogram in Seaborn for variables x and 'hue' y 
seaborn.histplot(data=penguins, x='bill_length_mm', hue='species', multiple='stack')
pyplot.show()
```

### Question 6

```python
# create a heatmap in Seaborn for a variable x and y
seaborn.heatmap(data=penguins.corr(numeric_only=True), annot=True)
pyplot.show()
```

### Question 7

```python
# create a box plot in Seaborn for variables x and y
seaborn.boxplot(data=penguins, x='species', y='bill_length_mm')
pyplot.show()
```

### Question 8

```python
# create a violin plot in Seaborn for variables x and y
seaborn.violinplot(data=penguins, x='species', y='bill_length_mm')
pyplot.show()
```

### Question 9

```python
# create a bar plot in Seaborn for variables x and y
seaborn.barplot(data=penguins, x='species', y='bill_length_mm')
pyplot.show()
```

### Question 10

```python
# add a title to a Seaborn plot assigned to the variable ax with the title 'My Title'
ax = seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
ax.set_title('My Title')
pyplot.show()
```

### Question 11

```python
# save a Seaborn plot ax to a file named 'output.png'
ax = seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
ax.set_title('My Title')
pyplot.savefig('output.png')
```

### Question 12

```python
# set the x and y axis labels to 'X Label' and 'Y Label' respectively for a Seaborn plot ax
ax = seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
pyplot.show()
```

### Question 13

```python
# change the color palette of a Seaborn plot to 'Blues'
seaborn.set_palette('Blues')
seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
pyplot.show()
```

### Question 14

```python
# display a Seaborn plot with a logarithmic scale on the y-axis
seaborn.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
pyplot.yscale('log')
pyplot.show()
```

### Question 15

```python
# add a regression line to a scatter plot in Seaborn for variables x and y
seaborn.regplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
pyplot.show()
```
